### Descripcion de las variables 

- business_id: Identificador del negocio. Esta variable es importante para realizar cualquier tipo de análisis que requiera identificar de manera única cada negocio en el conjunto de datos.

- name: Nombre del negocio. El nombre puede ser útil para identificar rápidamente el tipo de negocio o marca asociada, lo cual puede ser relevante en análisis de marca o reconocimiento de nombres populares.

- address: Dirección del negocio. La dirección puede ser relevante para realizar análisis geográficos, como la distribución de negocios por ubicación o la identificación de patrones espaciales.

- city: Ciudad donde se encuentra el negocio. La información de la ciudad puede ser útil para analizar la distribución de negocios por área geográfica o realizar comparaciones entre diferentes ciudades.

- state: Código de estado. Esta variable puede ser importante para analizar la distribución de negocios por estado y realizar análisis específicos por región geográfica.

- postal_code: Código postal del negocio. El código postal puede ser útil para realizar análisis a nivel local o para agrupar negocios por áreas específicas.

- latitude: Latitud geográfica de la ubicación del negocio. La latitud es una variable clave en análisis espaciales, ya que permite ubicar geográficamente cada negocio en un mapa.

- longitude: Longitud geográfica de la ubicación del negocio. Al igual que la latitud, la longitud es fundamental para ubicar geográficamente cada negocio en un mapa y realizar análisis espaciales.

- stars: Calificación del negocio en forma de estrellas. Esta variable puede ser utilizada para analizar la satisfacción de los clientes y comparar la calidad de diferentes negocios.

- review_count: Cantidad de reseñas del negocio. La cantidad de reseñas puede ser un indicador de la popularidad y la interacción de los clientes con el negocio, lo cual puede ser relevante para análisis de reputación o preferencias de los consumidores.

- is_open: Indicador de si el negocio está abierto (1) o cerrado (0). Esta variable puede ser utilizada para filtrar los negocios abiertos y realizar análisis específicos sobre ellos.

- attributes: Atributos adicionales del negocio. Los atributos pueden contener información relevante sobre características específicas del negocio, como aceptación de tarjetas de crédito, estacionamiento disponible, etc. Estos atributos pueden ser utilizados para realizar análisis detallados y segmentar los negocios según diferentes criterios.

- categories: Categorías a las que pertenece el negocio. Las categorías pueden ser utilizadas para clasificar los negocios en diferentes sectores o industrias, lo cual puede ser útil para realizar análisis sectoriales o comparar diferentes tipos de negocios.

- hours: Horario de funcionamiento del negocio. Esta variable puede ser importante para realizar análisis de horarios de mayor actividad, identificar patrones de apertura y cierre, o determinar la disponibilidad de servicios en momentos específicos del día.

### Anlisis de la columna state

```python

state_stars = df.groupby('state')['stars'].mean().reset_index()


state_stars = state_stars.sort_values('stars', ascending=False)


plt.figure(figsize=(12, 6))
plt.bar(state_stars['state'], state_stars['stars'])
plt.xlabel('Estado')
plt.ylabel('Promedio de estrellas')
plt.title('Relación entre estados y promedio de estrellas de los negocios')
plt.xticks(rotation=90)
plt.show()
```

```python

state_stars = df.groupby('state')['stars'].mean().reset_index()


average_stars_total = state_stars['stars'].mean()

print("Promedio de calificaciones 'stars' en todos los estados: {:.2f}".format(
    average_stars_total))
```

Con esta grafica, podemos visualizar gráficamente el promedio de calificaciones por estado en cervecerias. A simplre vista podemos ver que la mayoria de estados tiene  un promedio que ronda las 3.85 estrellas
